```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
#file_path = '/path/to/your/file.csv'  # Update this to the path where your CSV file is located
data = pd.read_csv('sheet 2.csv')

# Count the number of countries in each income group
income_group_counts = data['IncomeGroup'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
income_group_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Countries by Income Group')
plt.xlabel('Income Group')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('sheet 2.csv')

# Create a bar plot to show the number of countries in each income group
plt.figure(figsize=(14, 8))
sns.countplot(y='IncomeGroup', data=data, palette='viridis')
plt.title('Number of Countries by Income Group')
plt.xlabel('Number of Countries')
plt.ylabel('Income Group')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
#file_path = '/path/to/your/file.csv'  # Update this to the path where your CSV file is located
data = pd.read_csv('sheet 2.csv')

# Prepare the data for plotting
income_groups = data.groupby('IncomeGroup')['TableName'].apply(list)

# Plotting the grouped bar chart
fig, ax = plt.subplots(figsize=(20, 12))
for i, (group, countries) in enumerate(income_groups.items()):
    ax.barh([f'{group} ({len(countries)})']*len(countries), countries, label=group)

plt.title('Countries by Income Group')
plt.xlabel('Countries')
plt.ylabel('Income Group')
plt.grid(axis='x', linestyle='--', alpha=0.1)
plt.legend(title='Income Group')
plt.tight_layout()

# Show the plot
plt.show()
```

###  Top Countries by Population in 2023

```python
# Get population data for 2022
data_2022 = data[['Country Name', '2022']].sort_values(by='2022', ascending=False).head(10)

# Plot top countries by population in 2023
plt.figure(figsize=(12, 6))
plt.barh(data_2022['Country Name'], data_2022['2022'], color='skyblue')
plt.title('Top 10 Countries by Population in 2022')
plt.xlabel('Population')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```
